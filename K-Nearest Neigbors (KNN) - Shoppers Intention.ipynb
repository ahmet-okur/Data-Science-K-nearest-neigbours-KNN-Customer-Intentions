#### Data Augmentation
* to generate more Input data for training by using upsampling

```python
shoppers_df['Revenue'].value_counts()
```

```python
# Resample using Bootstrapping Method to generate random sample for each class
from sklearn.utils import resample
df_0 = shoppers_df[shoppers_df['Revenue']==False]
df_1 = shoppers_df[shoppers_df['Revenue']==True]
```

```python
# Apply Resample 
df_1_upsample = resample(df_1, n_samples = 8000, replace = True, random_state = 123)
```

```python
# Merge new samples with df_0 
shoppers_df2 = pd.concat([df_0, df_1_upsample])
```

```python
sns.catplot(x = 'Revenue', kind = 'count', data = shoppers_df2, height = 6, palette = 'magma')
```

### Data Pre-Processing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Laber encoder to convert catagorical data to numerical data becuse of to compute the distance on K-NN algo.
le = LabelEncoder()
le.fit(shoppers_df2['VisitorType'].drop_duplicates())
```

```python
shoppers_df2['VisitorType'] = le.transform(shoppers_df2['VisitorType'])
```

```python
shoppers_df2['VisitorType']
```

```python
le = LabelEncoder()
le.fit(shoppers_df2['Month'].drop_duplicates())
```

```python
shoppers_df2['Month'] = le.transform(shoppers_df2['Month'])
```

```python
shoppers_df2['Month']
```

### Features

```python
# we dropped the Bouncerates instead of ExitRates because we look for magnitude
X = shoppers_df2.drop(['ProductRelated', 'BounceRates', 'Revenue'], axis = 1)
Y = shoppers_df2['Revenue']
```

#### Using training and testing datasets to train the model

```python
# Split the dataset to test and train the model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X ,Y, test_size = 0.25, random_state = 3)
```

### KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# n_neighbours = Hyperparameters
knn = KNeighborsClassifier(n_neighbors = 15)
```

```python
knn.fit(x_train, y_train)
```

```python
# Accuracy is incresing by taking smallaer number of neighbors
knn.score(x_test, y_test) 
```

### Grid Seach CV
* To find the best hyperparameter values

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors' : np.arange(3, 60)}
```

```python
# n_neighbours = Hyperparameters
knn = KNeighborsClassifier()
# k-fold : 7 folds = Splitting Data into Train & Test 
knn_cv = GridSearchCV(knn, param_grid, cv = 7)
knn_cv.fit(x_train, y_train)
```

```python
knn_cv.best_params_
```

```python
knn_cv.best_score_
```

### Final KNN Model withn_neigbors = 3

```python
# n_neighbours = Hyperparameters
# minumun number of neigbor must be n+1, where n is number of class
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

### Clasification Metrics

```python
# yhat = predicted output
y_pred = knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, fmt = '0.0f', cmap = 'magma')
# Diagonals are true predictions
```

```python
print(classification_report(y_test, y_pred))
```
